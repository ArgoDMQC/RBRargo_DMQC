```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from correct_sqrt_error import correct_sqrt_error

# Load data
data = pd.read_csv('testdata.csv', header = 0, names=('PRES','CNDC','TEMP','PSALgood','PSALbad'))
Sgood = correct_sqrt_error(data['PSALbad'].values)

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.scatter(data['PSALgood'], data['PSALbad'] - data['PSALgood'], s=50, c=data['TEMP'], cmap='viridis')
plt.grid(True, which='both', linestyle='--')
plt.xlim(20, 40)
plt.ylim(-.5e-3, 2e-3)
plt.xlabel('Salinity')
plt.ylabel('Salinity error')
plt.colorbar(label='Temperature [°C]')
plt.title('Before correction')

plt.subplot(1, 2, 2)
plt.scatter(data['PSALgood'], Sgood - data['PSALgood'], s=50, c=data['TEMP'], cmap='viridis')
plt.grid(True, which='both', linestyle='--')
plt.xlim(20, 40)
plt.ylim(-.5e-3, 2e-3)
plt.xlabel('Salinity')
plt.ylabel('Salinity error')
plt.colorbar(label='Temperature [°C]')
plt.title('After correction')

plt.tight_layout()
plt.show()
```
